```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 1. Random Plots


x = np.arange(40)
data1 = np.random.randint(0, 101, 40)
data2 = np.random.randint(0, 101, 40)
data3 = np.random.randint(0, 101, 40)

fig, (ax1, ax2) = plt.subplots(2, 1)

ax1.plot(x, data1, color='orange', linewidth=10)
ax1.plot(x, data2, color='red', linestyle='--')
ax1.set_title('Top Plot')

ax2.scatter(x, data3, color='m', marker='D')
ax2.set_title('Bottom Plot')

plt.tight_layout()
plt.show()
```

```python
# 2. Van der Waals Gas

a = 16.02
b = 0.1124
R = 0.083144
n = 1

pressure_range = np.linspace(1, 10, 100)  
volume_range = np.linspace(10, 30, 100)

P, V = np.meshgrid(pressure_range, volume_range)

T = ((P + a * (n**2 / V**2)) * (V - n * b)) / (n * R)

plt.figure(figsize=(10, 6))
plt.pcolormesh(P, V, T, cmap='spring', vmin=np.min(T), vmax=np.max(T))
plt.colorbar(label='Temperature (K)')
plt.xlabel('Pressure (bar)')
plt.ylabel('Volume (L)')
plt.title('Temperature of Acetone (1 mole)')
plt.show()
```

```python
# 3. Monte Carlo 

def estimate_pi(N):
    points = np.random.rand(N, 2)
    distances = np.linalg.norm(points, axis=1)
    inside_circle = np.sum(distances <= 1)
    pi_estimate = 4 * inside_circle / N
    return pi_estimate
N_values = [int(1e3), int(1e4), int(1e5), int(1e6)]
pi_estimates = [estimate_pi(N) for N in N_values]
for N, pi_estimate in zip(N_values, pi_estimates):
    print(f"N = {N}:   {pi_estimate}")

N = int(1e4)
points = np.random.rand(N, 2)
distances = np.linalg.norm(points, axis=1)
inside_circle = distances <= 1

plt.figure(figsize=(6, 6))
plt.scatter(points[:, 0][inside_circle], points[:, 1][inside_circle], color='blue', label='Inside quarter circle')
plt.scatter(points[:, 0][~inside_circle], points[:, 1][~inside_circle], color='red', label='Outside quarter circle')
plt.axis('equal')
plt.text(0.05, 0.95, f'  {pi_estimates[1]}', transform=plt.gca().transAxes, fontsize=12, verticalalignment='top')
plt.title('Monte Carlo Estimation of  (N = 1e4)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
